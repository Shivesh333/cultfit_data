## Quality Check

```python
def Quality_check(df):
    d =dict(df.isnull().sum())
    for x in d:
        if d[x] == 0:
            print(f'No null value Found in {x}')
            print(f'datatype {df[x].dtype}')
        else:
            print(f'Null value found in {x}')
            print(f'datatype {df[x].dtype}')
```

```python
Quality_check(df)
```
